```python
# De maneira objetiva esse comando nos mostra as colunas que h√° valores faltantes , sendo  FALSE para as COLUNAS que N√ÉO tem valores faltantes e TRUE para os que TEM valores faltantes
df.isna().any()
```

```python
# Observando por meio de  de uma interface gr√°fica de maneira precisa onde  h√° os valores faltantes
#Obs: Os espa√ßos em branco correspondem aos valores faltantes
missingno.matrix(df)
```

```python
# Aqui ele vai me retornar todas as linhas que possui pelo menos um valor faltante (isna().any)
df[df.isna().any(axis = 1)] 
```

```python
# Me retorna linhas  que n√£o tem nenhum valor faltante, observe que agora = notna(). all
# em suma elimina os n/as

df[df.notna().all(axis = 1)]  
```

```python
# Aqui com o comando reset_index, eu reseto os valores naturais da ordem do data frame e mantenho apenas as amostras que est√£o presentes em todas as linhas e colunas
# OBS: Deixei de lado/joguei  fora todos os dados faltantes

df2 = df[df.notna().all(axis = 1)]. reset_index(drop = True)
```

```python
# Observe que todas as colunas possuem a mesma quantidade de amostras 
df2.info()
```

```python
# Observando graficamente (2), note que n√£o h√° lacunas, tudo preenchidos sem espa√ßos em branco
missingno.matrix(df2)
```

```python
# Aqui calculamos a porcentagem por meio do la√ßo de repeti√ß√£o (for) dos valores faltantes.
for col in df.columns:
    perc = np.mean(df[col].isna())
    print (f'{col}: {round(perc*100)} %')
```

```python
# Selecionei a coluna total vacinations e dentro dela s√≥ quero os que s√£o menores ou iguais a dez mil

dff = df[df['total_vaccinations'] <=10000]
dff.info()
```

```python
# Esta fun√ß√£o √© an√°loga a de cima, a partir de ent√£o irei usa-la.

dff2 = df.query('total_vaccinations <=10000')
dff2
```

```python
# Na coluna country eu selecionei o Brazil e na coluna people vaccinated os que est√£o acima de 100000 (cem mil)
dff3 = df.query('country == "Brazil" & people_vaccinated >= 100000')
dff3
```

```python
# OU no formato de  | , aqui ele ir√° selecionar todos as linhas que possuem mais de 1 mih√ß√£o na coluna total vaccinations ou as linhas que tem mais de cem mil na coluna people fully vaccination
dff4 = df.query ('total_vaccinations > 1e6 | people_fully_vaccinated > 1e5')
dff4
```

```python
# Filtrando data espec√≠fica e na√ß√£o
df.query('date >= " 2021-06-01" & country == "Afghanistan"')
```

```python
# FILTRANDO INTERVALOS DE TEMPO entre dia 01 e dia 31
df.query ('date in ["2021-05-01", "2021-05-31"]')
```

```python
# ELIMINANDO as colunas que possuem  N/AS nas colunas
dff5= df.dropna(axis=1)
dff5
```

```python
# Substituindo N/AS pela mediana
#Calculado a mediana
dfm= df.median()
dfm
```

```python
# listando os valores
mediana = df.median().values
mediana
```

```python
#Selecionando as colunas com valores num√©ricos
cols = ['total_vaccinations', 'people_vaccinated', 'people_fully_vaccinated', 'daily_vaccinations_raw', 'daily_vaccinations', 'total_vaccinations_per_hundred','people_vaccinated_per_hundred','people_fully_vaccinated_per_hundred','daily_vaccinations_per_million']
len(cols)
```

```python
#substituindo os valores faltantes pela mediana de cada coluna
for i in range(len(mediana)):
    df[cols[i]] = df[cols[i]].fillna(mediana[i])
```

```python
#Conferindo, todos as colunas possuem a mesma quantidade de 
df.info()
```
